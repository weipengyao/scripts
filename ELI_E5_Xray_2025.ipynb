```python
thickness_SiN     = np.array([30.0, 100.0, 150.0, 300.0])
cutoff_energy_SiN = np.array([ 3.2,   3.2,   5.7,   4.6])

thickness_Al      = np.array([2000.0, 3000.0, 20000.0, 50000.0, 100000.0])
cutoff_energy_Al  = np.array([  27.6,   18.7,    16.8,     5.7,      1.1])

thickness_Au      = np.array([2000.0, 10000.0, 25000, 50000.0, 125000.0])
cutoff_energy_Au  = np.array([   5.7,     4.6,   1.1,     1.1,       1.1])
```

```python
# Set font to Calibri and increase font sizes
plt.rcParams.update({
    'font.family': 'Calibri',
    'font.size': 24,
    'axes.titlesize': 28,
    'axes.labelsize': 24,
    'legend.fontsize': 20,
    'xtick.labelsize': 24,
    'ytick.labelsize': 24
})

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(thickness_SiN, cutoff_energy_SiN, marker='o', label="SiN")
plt.plot(thickness_Al,  cutoff_energy_Al,  marker='*', linestyle='--', label="Al")
plt.plot(thickness_Au,  cutoff_energy_Au,  marker='s', linestyle=':', label="Au")
plt.title("thickness vs cutoff energy")
plt.xlabel("Thickness (nm)")
plt.ylabel("Cutoff Energy (MeV)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xscale('log')
plt.xlim(10, 150000)
plt.ylim(0, 30)
plt.grid(True, which='both', linestyle='-', linewidth=0.5, alpha=0.5)
plt.show()
```
