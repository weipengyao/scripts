## Scientific background

Accurate evaluation of the Coulomb logarithm is essential in determining transport properties—such as diffusion, viscosity, resistivity, and temperature relaxation—in plasmas. Under the usual weakly coupled, hot and dilute conditions, the Coulomb logarithm, $\ln \Lambda$, arises from integrating the impact parameter between Debye-screened and closest-approach limits—yielding values typically between 5 and 15—but this treatment assumes large plasma parameters and becomes unreliable in dense, cold regimes. In stiffly coupled plasmas where the coupling parameter $\Gamma \gtrsim 1$, traditional Landau-Spitzer approaches fail as correlations become significant; here, the binary collision picture must be extended through effective potentials that include correlation and screening effects, as shown in effective-potential transport theories. In these strongly coupled, cold and dense regimes, the standard Coulomb logarithm may even become negative or poorly defined, leading to unphysical predictions. Such failures can significantly skew estimates of energy relaxation and transport timescales.

Therefore, a better calculation of the Coulomb logarithm—one that properly accounts for screening, quantum effects (when de Broglie wavelengths approach screening lengths), and particle correlations—is critical for correctly modeling transport and relaxation phenomena in cold, dense plasmas such as those found in inertial confinement fusion, ultracold neutral plasmas, astrophysical interiors, and warm dense matter contexts.

Here, based on the reference paper cited above, a handy equation which includes the correction introduced in the reference above is as follows:

$\begin{gathered}\Theta=\ln (1+0.7 \Lambda), \\ \Lambda=\frac{b_{\max }}{b_{\min }}, \\ b_{\max }=\min \left[\max \left(\lambda_D, R_i\right), \sqrt{\frac{T_e}{m_e \omega_0^2}}\right] \\ b_{\min }=\min \left[\max \left(\frac{Z_i q_e^2}{3 T_e \epsilon_0}, \frac{\hbar}{2 \sqrt{3 m_e T_e}}\right), R_i\right]\end{gathered}$

where $R_i$, $q_e$, and $\hbar$ are the averaged inter-ion distance, the unsigned electron charge, and reduced Planck constant, respectively.

## Scientific background

Laser-plasma interaction is a cornerstone of both fundamental plasma physics research and practical applications ranging from inertial confinement fusion (ICF) and particle acceleration to high-energy-density physics experiments and advanced light sources. In these regimes, the electron temperature ($T_e$) plays a pivotal role in determining key plasma parameters—such as the Debye length, ion-acoustic speed, collisionality, and instability growth rates—that govern energy transport, wave propagation, and laser energy coupling efficiency.

However, direct measurement of $T_e$ in intense laser-plasma environments is notoriously challenging: diagnostics like Thomson scattering, X-ray spectroscopy, or bremsstrahlung analysis are often hampered by the transient, non-uniform, and non-equilibrium nature of the plasma, as well as by strong electromagnetic noise and geometric constraints. As a result, $T_e$ is frequently inferred indirectly from models or scaling laws that may neglect kinetic effects, turbulence, or multi-species interactions, leading to substantial uncertainties.

 A more accurate estimate of $T_e$ for given laser intensity and plasma density is therefore critical — not only for correctly predicting the onset and evolution of laser-driven instabilities, but also for optimizing energy deposition, minimizing deleterious scattering and filamentation, and improving the predictive power of both hydrodynamic and kinetic simulation frameworks.

Under the above-mentioned regime (i.e., low temperature and low density experiment with $T_e \le 250$ eV and $n_e/n_c < 0.01$), we can safely neglect the electron thermal diffusion and accounting only for the inverse Bremsstrahlung laser absorption calculated on the transversely averaged intensity. The resulting electron  temperature evolution follows the equation below:

$(3 / 2) n_e d_t T_e=\nu_B I_0 / q_e$

where $\nu_B$ is the bremsstrahlung laser absorption coefficient, $I_0$ is the laser intensity (usually with a Gaussiann temporal profile), and $q_e$ is the elementary electron charge.

````python
import numpy as np

def integrate_Te_evolution(
    Te0,             # a list with initial temperature [eV]
    ne,              # electron number density [cm^-3]
    I_laser,         # laser intensity array [W/cm^2]
    nu_B_func,       # function: nu_B(temp, coul_alog_func, Z0, ne, omg_ts, omg_pe)
    coul_alog_func,  # function: coul_alog(temp, ne, Z0, omg_ts)
    d_time           # time step [s]
):
    """
    Integrate electron temperature over time using inverse Bremsstrahlung heating.

    Returns:
        Te_list: list of electron temperatures [eV] for each time step
    """
    Te_list = Te0
    num_steps = len(I_laser)

    for t in range(num_steps):
        # (3/2) n_e dT_e/dt = ν_B * I / q_e
        dt = (nu_B_func(Te_list[t], coul_alog_func, Z0, ne, omg_ts, omg_pe) * I_laser[t] / (1.5 * ne)) \
             / 1.6e-19 * d_time
        Te_list.append(Te_list[t] + dt)

    return Te_list

def constants():
    # constants in CGS
    c    = 3e10            # cm/s, light speed
    qe   = 4.8e-10         # elementary charge
    hbar = 1.0e-27         # Planck constant
    me   = 9.1e-28         # electron mass, g
    return c, qe, hbar, me

c, qe, hbar, me = constants()

# define the laser
def setup_laser(E_ts, tau_ts, d_ts, lmd_ts):
    '''
    E_ts   = 110.0   # Laser energy (J)
    tau_ts = 5.0e-9  # Laser duration (s)
    d_ts   = 0.1     # Laser focal spot diameter (cm)
    lmd_ts = 1.0*1e-4 # cm, 1 omega laser (1 um)
    '''
    S_ts   = np.pi * (d_ts/2.)**2 # cm^2
    I_ts   = E_ts / tau_ts / S_ts # W/cm2
    # print('Laser intensity for TS = {:.2e} W/cm2'.format(I_ts))
    omg_ts = 2.*np.pi*c/lmd_ts
    # print('Laser angular frequency for TS = {:.2e} rad/s'.format(omg_ts))
    return E_ts, tau_ts, d_ts, lmd_ts, S_ts, I_ts, omg_ts

E_ts, tau_ts, d_ts, lmd_ts, S_ts, I_ts, omg_ts = setup_laser(110.0, 5.0e-9, 0.1, 1.0*1e-4)

# setup initial condition
def initial_condition(Te, Z0, ne, t_end):
    # Te = [0.1]            # initial temperature
    # Z0 = 1.               # plasma material
    # ne = 2.0e17           # electron number density cm^-3
    # t_end = 10e-9         # estimation time
    omg_pe  = 5.64e4 * ne**0.5
    # print('plasma electron angular frequency = {:.2e} rad/s'.format(omg_pe))
    return Te, Z0, ne, t_end, omg_pe

Te, Z0, ne, t_end, omg_pe = initial_condition([0.1], 1., 1.0e18, 10e-9)


# create the array for time evolution
def time_evo(num, t_end):
    time = np.linspace(0,t_end,num)
    d_time = time[1] - time[0]
    return time, d_time

time, d_time = time_evo(1000, t_end)

# Gaussian profile for the Laser time-evolution
def tgaussian(fwhm,order,center):
    '''
    fwhm = tau_ts         # equals to tau_ts above
    order = 4.            # order of gaussian profile
    center = t_end / 2.0  # center of gaussian profile
    '''
    sigma = (0.5*fwhm)**order/np.log(2.)
    return np.exp(-(time-center)**order / sigma)

I_laser = I_ts*tgaussian(tau_ts,4.,t_end / 2.0)

# calculation of the Coulomb logarithm
def LMD2(temp, ne, Z0, omg_ts):
    Ld = 7.43e2 * temp**0.5 * ne**(-0.5)  # cm
    ni = ne/Z0  # cm^-3
    Ri = ni**(-1./3.)  # cm
    vte  = 4.19e7 * temp**0.5              # cm/s
    bmax = np.minimum(np.maximum(Ld, Ri), vte/omg_ts)
    bmin = np.minimum(np.maximum(Z0*qe**2 / (temp*1.6e-12), hbar/np.sqrt(me*temp*1.6e-12)), Ri)  # bug from SI->CGS ???
    return np.log(1 + 0.7 * bmax/bmin)

# Inverse bremsstrahlung absorption coefficient for radiation of angular frequency 𝜔:
# Considered the temperature dependence
def nu_B(temp,coul_alog, Z0, ne, omg_ts, omg_pe):
    return 3.1e-7 * Z0 * ne**2 * coul_alog(temp, ne, Z0, omg_ts) * temp**(-1.5) * omg_ts**-2. * (1.-omg_pe**2/omg_ts**2)**(-0.5)   # cm^-1

Te = integrate_Te_evolution(
    Te0=Te,
    ne=ne,
    I_laser=I_laser,
    nu_B_func=nu_B,
    coul_alog_func=LMD2,
    d_time=d_time
)

# print('final Te = {:.1f} eV'.format(Te[-1]))


def test_case_1():
    c, qe, hbar, me = constants()
    Te, Z0, ne, t_end, omg_pe = initial_condition([0.1], 1., 2.0e17, 10e-9)
    # define the laser
    E_ts, tau_ts, d_ts, lmd_ts, S_ts, I_ts, omg_ts = setup_laser(110.0, 5.0e-9, 0.1, 1.0*1e-4)
    time, d_time = time_evo(1000, t_end)
    expected = 31.89944740223177
    output   = integrate_Te_evolution(Te,ne,I_laser,nu_B,LMD2,d_time)[-1]
    assert np.allclose(output, expected), f"Fail: got {output}, expected {expected}"
    print("Pass ✅: output matches expected value")
test_case_1()

def test_case_2():
    c, qe, hbar, me = constants()
    Te, Z0, ne, t_end, omg_pe = initial_condition([0.1], 1., 1.0e18, 10e-9)
    # define the laser
    E_ts, tau_ts, d_ts, lmd_ts, S_ts, I_ts, omg_ts = setup_laser(110.0, 5.0e-9, 0.1, 1.0*1e-4)
    time, d_time = time_evo(1000, t_end)
    expected = 70.05553494626747
    output   = integrate_Te_evolution(Te,ne,I_laser,nu_B,LMD2,d_time)[-1]
    assert np.allclose(output, expected), f"Fail: got {output[-1]}, expected {expected}"
    print("Pass ✅: output matches expected value")
test_case_2()

def test_case_3():
    c, qe, hbar, me = constants()
    Te, Z0, ne, t_end, omg_pe = initial_condition([0.1], 1., 1.0e18, 10e-9)
    # define the laser
    E_ts, tau_ts, d_ts, lmd_ts, S_ts, I_ts, omg_ts = setup_laser(110.0, 2.0e-9, 0.1, 1.0*1e-4)
    time, d_time = time_evo(1000, t_end)
    expected = 70.05553494626747
    output   = integrate_Te_evolution(Te,ne,I_laser,nu_B,LMD2,d_time)[-1]
    assert np.allclose(output, expected), f"Fail: got {output[-1]}, expected {expected}"
    print("Pass ✅: output matches expected value")
test_case_3()
````
