```python
# units

me = 9.10938356e-31  # kg
qe = 1.602176634e-19  # C
eps0 = 8.854187817e-12  # F/m
c = 3e8  # m/s

lmd = 0.8 * 1e-6  # m
wr = 2.0 * np.pi * c / lmd

Er = me * wr * c / qe
nr = eps0 * me * wr**2 / qe**2

def get_S_para_t(S, label=None):
    Ey = np.array(S.Probe(2,'Ey',units=["um", "fs"]).getData())
    S_para_t = np.zeros(Ey.shape[0])
    for i in range(Ey.shape[0]):
        Ey = np.array(S.Probe(2,'Ey',units=["um", "fs"
                                            # , "V/m"
                                            ]).getData()[i])
        rho_e = np.array(S.Probe(2,'Rho_eon',units=["um", "fs"
                                                    # , 'C m^{-3}'
                                                    ]).getData()[i])
        # S_para_t[i] = np.sum(rho_e/nr/qe * (-1.0) * np.abs(Ey/Er)**2) / lmd
        S_para_t[i] = np.sum(rho_e * (-1.0) * np.abs(Ey)**2)

    time = np.array(S.Probe(2,'Ey',units=["um", "fs"]).getTimes())
    return time, S_para_t, label
```

```python
time0, S_para_t0, label0 = get_S_para_t(S0, 'a0=282')
time1, S_para_t1, label1 = get_S_para_t(S1, 'a0=353')
time2, S_para_t2, label2 = get_S_para_t(S2, 'a0=424')
time3, S_para_t3, label3 = get_S_para_t(S3, 'a0=750')
```

```python
plt.plot(time0, S_para_t0, label=label0)
plt.plot(time1, S_para_t1, label=label1)
plt.plot(time2, S_para_t2, label=label2)
plt.plot(time3, S_para_t3, label=label3)

plt.xlabel("time (fs)")
plt.ylabel("S_para (NOT normalized)")
plt.legend()
plt.xlim([0, 100])
plt.ylim([0, 3e9])
# plt.title(r"$S=n_c^{-1} \lambda_L^{-1} \int_0^{L_x} dx n_e |a_0|^2$")
plt.title(r"$S=\int_0^{L_x} dx \rho_e |E_y|^2$")

plt.savefig("/Users/yao/Desktop/S_para_t.png", dpi=300)
```
