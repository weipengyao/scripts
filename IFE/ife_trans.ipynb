```python
# instead of finding the maximum Ex field position, we can also 
# check the density peak position, which will be less messy.

import numpy as np
import matplotlib.pyplot as plt

try:
    from scipy.signal import find_peaks, savgol_filter
    SCIPY_AVAILABLE = True
except ImportError:
    SCIPY_AVAILABLE = False


def _moving_average(y, window=5):
    """Simple moving average with edge padding."""
    if window <= 1:
        return y.copy()
    pad = window // 2
    ypad = np.pad(y, (pad, pad), mode="edge")
    kernel = np.ones(window) / window
    return np.convolve(ypad, kernel, mode="valid")


def _smooth_1d(y, method="savgol", window=9, polyorder=2):
    """Smooth a 1D signal."""
    y = np.asarray(y)
    if window <= 1:
        return y.copy()

    if method == "savgol" and SCIPY_AVAILABLE:
        # window must be odd and <= len(y)
        if window % 2 == 0:
            window += 1
        window = min(window, len(y) if len(y) % 2 == 1 else len(y) - 1)
        if window < polyorder + 2:
            return y.copy()
        return savgol_filter(y, window_length=window, polyorder=polyorder)
    else:
        return _moving_average(y, window=window)


def track_leftmost_peak(
    ne,
    x,
    t,
    smooth=True,
    smooth_method="savgol",
    smooth_window=9,
    polyorder=2,
    min_height_frac=0.05,
    prominence_frac=0.03,
    continuity=True,
    max_jump=None,
    fallback="nan",
    plot=True,
    velocity_smooth=False,
    velocity_smooth_window=7,
    figsize=(7, 5),
):
    """
    Track the leftmost significant density peak in a 2D array ne(t, x).

    Parameters
    ----------
    ne : array-like, shape (nt, nx)
        Density array.
    x : array-like, shape (nx,)
        Spatial axis in um.
    t : array-like, shape (nt,)
        Time axis in fs.
    smooth : bool
        Whether to smooth each spatial profile before peak detection.
    smooth_method : str
        "savgol" or "moving_average".
    smooth_window : int
        Smoothing window in x direction.
    polyorder : int
        Polynomial order for Savitzky-Golay.
    min_height_frac : float
        Minimum peak height as fraction of the frame maximum.
    prominence_frac : float
        Minimum peak prominence as fraction of the frame maximum.
    continuity : bool
        If True, enforce temporal continuity by choosing the peak closest
        to the previous one among acceptable candidates.
    max_jump : float or None
        Maximum allowed jump in x (same unit as x) between consecutive frames.
        If None, no explicit jump limit is applied.
    fallback : str
        What to do if no peak is found: "nan" or "previous".
    plot : bool
        If True, plot x_peak(t) and v(t).
    velocity_smooth : bool
        If True, smooth x_peak(t) before differentiating.
    velocity_smooth_window : int
        Window for smoothing x_peak(t) before velocity calculation.
    figsize : tuple
        Figure size.

    Returns
    -------
    result : dict
        {
            "t": t,
            "x_peak": x_peak,
            "v_peak": v_peak,
            "peak_index": peak_index
        }
        where x_peak is in um and v_peak is in um/fs.
    """
    ne = np.asarray(ne)
    x = np.asarray(x)
    t = np.asarray(t)

    if ne.ndim != 2:
        raise ValueError("ne must have shape (nt, nx)")
    if ne.shape[1] != len(x):
        raise ValueError("x length must match ne.shape[1]")
    if ne.shape[0] != len(t):
        raise ValueError("t length must match ne.shape[0]")

    nt, nx = ne.shape
    x_peak = np.full(nt, np.nan, dtype=float)
    peak_index = np.full(nt, -1, dtype=int)

    prev_x = None

    for it in range(nt):
        y = ne[it].astype(float)

        if smooth:
            y_use = _smooth_1d(
                y, method=smooth_method, window=smooth_window, polyorder=polyorder
            )
        else:
            y_use = y.copy()

        ymax = np.nanmax(y_use)
        if not np.isfinite(ymax) or ymax <= 0:
            if fallback == "previous" and prev_x is not None:
                x_peak[it] = prev_x
            continue

        min_height = min_height_frac * ymax
        min_prom = prominence_frac * ymax

        if SCIPY_AVAILABLE:
            peaks, props = find_peaks(y_use, height=min_height, prominence=min_prom)
        else:
            # Simple fallback local maxima finder
            mask = (y_use[1:-1] > y_use[:-2]) & (y_use[1:-1] >= y_use[2:]) & (y_use[1:-1] > min_height)
            peaks = np.where(mask)[0] + 1
            props = {}

        if len(peaks) == 0:
            if fallback == "previous" and prev_x is not None:
                x_peak[it] = prev_x
                peak_index[it] = np.argmin(np.abs(x - prev_x))
            continue

        # Sort candidates from left to right
        peaks = np.sort(peaks)
        x_candidates = x[peaks]

        if prev_x is None or not continuity:
            chosen_idx = peaks[0]  # leftmost significant peak
        else:
            # Prefer candidates not too far from previous peak
            distances = np.abs(x_candidates - prev_x)

            if max_jump is not None:
                valid = distances <= max_jump
                if np.any(valid):
                    chosen_idx = peaks[np.argmin(distances[valid])]
                    # careful: argmin on subset
                    valid_peaks = peaks[valid]
                    valid_dist = distances[valid]
                    chosen_idx = valid_peaks[np.argmin(valid_dist)]
                else:
                    # If nothing is close enough, fall back to leftmost or previous
                    if fallback == "previous":
                        x_peak[it] = prev_x
                        peak_index[it] = np.argmin(np.abs(x - prev_x))
                        continue
                    chosen_idx = peaks[0]
            else:
                # among detected peaks, choose the one closest to previous
                # but bias toward the left if several are close
                chosen_idx = peaks[np.argmin(distances)]

                # Optional safeguard:
                # if the closest peak is very far from prev_x, use leftmost instead
                # Comment this block out if undesired.
                if len(x) > 1:
                    dx = np.median(np.diff(x))
                    if np.abs(x[chosen_idx] - prev_x) > 10 * dx:
                        chosen_idx = peaks[0]

        x_peak[it] = x[chosen_idx]
        peak_index[it] = chosen_idx
        prev_x = x_peak[it]

    # Velocity
    x_for_v = x_peak.copy()
    if velocity_smooth:
        valid = np.isfinite(x_for_v)
        if np.sum(valid) >= max(5, velocity_smooth_window):
            x_for_v_valid = _smooth_1d(
                x_for_v[valid],
                method=smooth_method,
                window=velocity_smooth_window,
                polyorder=min(polyorder, 2),
            )
            x_for_v[valid] = x_for_v_valid

    v_peak = np.full_like(x_for_v, np.nan, dtype=float)
    valid = np.isfinite(x_for_v)

    if np.sum(valid) >= 2:
        v_peak[valid] = np.gradient(x_for_v[valid], t[valid])

    if plot:
        fig1, ax1 = plt.subplots(figsize=figsize)
        ax1.plot(t, x_peak, "o-", ms=3)
        ax1.set_xlabel("t (fs)")
        ax1.set_ylabel("Leftmost peak position (um)")
        ax1.set_title("Peak position vs time")
        ax1.grid(True, alpha=0.3)
        plt.tight_layout()

        fig2, ax2 = plt.subplots(figsize=figsize)
        ax2.plot(t, v_peak, "o-", ms=3)
        ax2.set_xlabel("t (fs)")
        ax2.set_ylabel("Peak velocity (um/fs)")
        ax2.set_title("Peak velocity vs time")
        ax2.grid(True, alpha=0.3)
        plt.tight_layout()

    return {
        "t": t,
        "x_peak": x_peak,
        "v_peak": v_peak,
        "peak_index": peak_index,
    }
```

```python
import numpy as np
import matplotlib.pyplot as plt

try:
    from scipy.signal import find_peaks, savgol_filter
    SCIPY_AVAILABLE = True
except ImportError:
    SCIPY_AVAILABLE = False


def _moving_average(y, window=5):
    if window <= 1:
        return y.copy()
    pad = window // 2
    ypad = np.pad(y, (pad, pad), mode="edge")
    kernel = np.ones(window) / window
    return np.convolve(ypad, kernel, mode="valid")


def _smooth_1d(y, method="savgol", window=9, polyorder=2):
    y = np.asarray(y)
    if window <= 1:
        return y.copy()

    if method == "savgol" and SCIPY_AVAILABLE:
        if window % 2 == 0:
            window += 1
        window = min(window, len(y) if len(y) % 2 == 1 else len(y) - 1)
        if window < polyorder + 2:
            return y.copy()
        return savgol_filter(y, window_length=window, polyorder=polyorder)
    else:
        return _moving_average(y, window=window)


def track_leftmost_peak(
    ne,
    x,
    t,
    smooth=True,
    smooth_method="savgol",
    smooth_window=9,
    polyorder=2,
    min_height_frac=0.05,
    prominence_frac=0.03,
    continuity=True,
    max_jump=None,
    fallback="nan",
    velocity_smooth=False,
    velocity_smooth_window=7,
):
    """
    Track the leftmost significant peak in ne(t, x).

    Parameters
    ----------
    ne : array, shape (nt, nx)
    x  : array, shape (nx,)
    t  : array, shape (nt,)

    Returns
    -------
    dict with keys:
        "t", "x_peak", "v_peak", "peak_index"
    """
    ne = np.asarray(ne)
    x = np.asarray(x)
    t = np.asarray(t)

    if ne.ndim != 2:
        raise ValueError("ne must have shape (nt, nx)")
    if ne.shape[1] != len(x):
        raise ValueError("x length must match ne.shape[1]")
    if ne.shape[0] != len(t):
        raise ValueError("t length must match ne.shape[0]")

    nt, nx = ne.shape
    x_peak = np.full(nt, np.nan, dtype=float)
    peak_index = np.full(nt, -1, dtype=int)

    prev_x = None

    for it in range(nt):
        y = ne[it].astype(float)

        if smooth:
            y_use = _smooth_1d(
                y, method=smooth_method, window=smooth_window, polyorder=polyorder
            )
        else:
            y_use = y.copy()

        ymax = np.nanmax(y_use)
        if not np.isfinite(ymax) or ymax <= 0:
            if fallback == "previous" and prev_x is not None:
                x_peak[it] = prev_x
            continue

        min_height = min_height_frac * ymax
        min_prom = prominence_frac * ymax

        if SCIPY_AVAILABLE:
            peaks, props = find_peaks(y_use, height=min_height, prominence=min_prom)
        else:
            mask = (
                (y_use[1:-1] > y_use[:-2]) &
                (y_use[1:-1] >= y_use[2:]) &
                (y_use[1:-1] > min_height)
            )
            peaks = np.where(mask)[0] + 1

        if len(peaks) == 0:
            if fallback == "previous" and prev_x is not None:
                x_peak[it] = prev_x
                peak_index[it] = np.argmin(np.abs(x - prev_x))
            continue

        peaks = np.sort(peaks)
        x_candidates = x[peaks]

        if prev_x is None or not continuity:
            chosen_idx = peaks[0]
        else:
            distances = np.abs(x_candidates - prev_x)

            if max_jump is not None:
                valid = distances <= max_jump
                if np.any(valid):
                    valid_peaks = peaks[valid]
                    valid_dist = distances[valid]
                    chosen_idx = valid_peaks[np.argmin(valid_dist)]
                else:
                    if fallback == "previous":
                        x_peak[it] = prev_x
                        peak_index[it] = np.argmin(np.abs(x - prev_x))
                        continue
                    chosen_idx = peaks[0]
            else:
                chosen_idx = peaks[np.argmin(distances)]

                if len(x) > 1:
                    dx = np.median(np.diff(x))
                    if np.abs(x[chosen_idx] - prev_x) > 10 * dx:
                        chosen_idx = peaks[0]

        x_peak[it] = x[chosen_idx]
        peak_index[it] = chosen_idx
        prev_x = x_peak[it]

    x_for_v = x_peak.copy()
    if velocity_smooth:
        valid = np.isfinite(x_for_v)
        if np.sum(valid) >= max(5, velocity_smooth_window):
            x_for_v_valid = _smooth_1d(
                x_for_v[valid],
                method=smooth_method,
                window=velocity_smooth_window,
                polyorder=min(polyorder, 2),
            )
            x_for_v[valid] = x_for_v_valid

    v_peak = np.full_like(x_for_v, np.nan, dtype=float)
    valid = np.isfinite(x_for_v)
    if np.sum(valid) >= 2:
        v_peak[valid] = np.gradient(x_for_v[valid], t[valid])

    return {
        "t": t,
        "x_peak": x_peak,
        "v_peak": v_peak,
        "peak_index": peak_index,
    }


def _get_smilei_axis(diag, axis_candidates=("x", "axis1")):
    """
    Try a few common Smilei axis names and return a 1D x-axis.
    """
    last_err = None
    for axname in axis_candidates:
        try:
            ax = np.array(diag.getAxis(axname))
            # common Smilei outputs: shape (nx, 2) or (nx,)
            if ax.ndim == 2:
                return ax[:, 0]
            return ax
        except Exception as e:
            last_err = e
    raise RuntimeError(f"Could not read x-axis from Smilei diagnostic. Last error: {last_err}")


def extract_ne_xt_from_smilei(
    S,
    diag_number=3,
    y_halfwidth=5,
    units=('um', 'fs', 'cm^-3'),
):
    """
    Extract ne(t, x), x, t from Smilei.
    """
    diag = S.ParticleBinning(
        diag_number,
        units=list(units),
        average={
            "z": "all",
            "y": [int(S.namelist.Ly / 2.) - y_halfwidth,
                  int(S.namelist.Ly / 2.) + y_halfwidth],
        },
    )

    ne = np.array(diag.getData())      # expected shape (nt, nx)
    x = _get_smilei_axis(diag)
    t = np.array(diag.getTimes())

    if ne.ndim != 2:
        raise ValueError(f"Expected ne to be 2D (nt, nx), but got shape {ne.shape}")

    if ne.shape[0] != len(t):
        raise ValueError(
            f"Mismatch: ne.shape[0]={ne.shape[0]} but len(t)={len(t)}"
        )
    if ne.shape[1] != len(x):
        raise ValueError(
            f"Mismatch: ne.shape[1]={ne.shape[1]} but len(x)={len(x)}"
        )

    return ne, x, t


def track_leftmost_peak_multiple_cases(
    S_list,
    a0_list,
    diag_number=3,
    y_halfwidth=5,
    units=('um', 'fs', 'cm^-3'),
    smooth=True,
    smooth_method="savgol",
    smooth_window=9,
    polyorder=2,
    min_height_frac=0.05,
    prominence_frac=0.03,
    continuity=True,
    max_jump=None,
    fallback="previous",
    velocity_smooth=True,
    velocity_smooth_window=7,
    figsize=(8, 6),
    save_prefix=None,
):
    """
    Run leftmost peak tracking for multiple Smilei cases and plot all curves together.

    Parameters
    ----------
    S_list : list
        List of Smilei simulation handles, e.g. [S0, S1, ..., S6]
    a0_list : list
        List of corresponding a0 values, e.g. [300, 400, ..., 900]

    Returns
    -------
    results : dict
        results[a0] = {
            "t": ...,
            "x_peak": ...,
            "v_peak": ...,
            "peak_index": ...,
            "x": ...,
            "ne": ...
        }
    """
    if len(S_list) != len(a0_list):
        raise ValueError("S_list and a0_list must have the same length")

    results = {}

    fig1, ax1 = plt.subplots(figsize=figsize)
    fig2, ax2 = plt.subplots(figsize=figsize)

    for S, a0 in zip(S_list, a0_list):
        ne, x, t = extract_ne_xt_from_smilei(
            S,
            diag_number=diag_number,
            y_halfwidth=y_halfwidth,
            units=units,
        )

        res = track_leftmost_peak(
            ne=ne,
            x=x,
            t=t,
            smooth=smooth,
            smooth_method=smooth_method,
            smooth_window=smooth_window,
            polyorder=polyorder,
            min_height_frac=min_height_frac,
            prominence_frac=prominence_frac,
            continuity=continuity,
            max_jump=max_jump,
            fallback=fallback,
            velocity_smooth=velocity_smooth,
            velocity_smooth_window=velocity_smooth_window,
        )

        results[a0] = {
            "t": res["t"],
            "x_peak": res["x_peak"],
            "v_peak": res["v_peak"],
            "peak_index": res["peak_index"],
            "x": x,
            "ne": ne,
        }

        ax1.plot(res["t"], res["x_peak"], "o-", ms=3, label=fr"$a_0={a0}$")
        ax2.plot(res["t"], res["v_peak"], "o-", ms=3, label=fr"$a_0={a0}$")

    ax1.set_xlabel("t (fs)")
    ax1.set_ylabel("Leftmost peak position (um)")
    ax1.set_title("Peak position vs time")
    ax1.grid(True, alpha=0.3)
    ax1.legend()

    ax2.set_xlabel("t (fs)")
    ax2.set_ylabel("Peak velocity (um/fs)")
    ax2.set_title("Peak velocity vs time")
    ax2.grid(True, alpha=0.3)
    ax2.legend()

    plt.figure(fig1.number)
    plt.tight_layout()

    plt.figure(fig2.number)
    plt.tight_layout()

    if save_prefix is not None:
        fig1.savefig(f"{save_prefix}_peak_position_vs_time.pdf", dpi=300, bbox_inches="tight")
        fig2.savefig(f"{save_prefix}_peak_velocity_vs_time.pdf", dpi=300, bbox_inches="tight")

    return results
```

```python
S_list  = [S0, S1, S2, S3, S4, S5, S6]
a0_list = [300, 400, 500, 600, 700, 800, 900]

results = track_leftmost_peak_multiple_cases(
    S_list=S_list,
    a0_list=a0_list,
    diag_number=3,
    y_halfwidth=5,
    smooth=True,
    smooth_window=11,
    min_height_frac=0.08,
    prominence_frac=0.04,
    continuity=True,
    max_jump=0.6,              # in um; tune if needed
    fallback="previous",
    velocity_smooth=True,
    velocity_smooth_window=7,
    save_prefix="ne_leftmost_peak_all_cases",
)
```

```python
# replace peak-tracking with front-tracking, which is more stable and less noisy. We can define the front as the position where the density first exceeds a certain threshold (e.g., 10% of the maximum density in that frame). This should give us a smoother curve for the front position over time, and we can still calculate the velocity by differentiating the front position.

"""
Track the density front (steepest positive gradient dn/dx) instead of the
density peak.  More robust for strong-laser cases where multiple peaks form.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter1d


# ── core tracker ────────────────────────────────────────────────────────────

def track_front(ne, x, t, smooth_pts=5, threshold_frac=0.1, max_jump=None):
    """
    At each timestep find the position of the largest positive dne/dx
    (i.e. the steepest rising edge = the shock / compression front).

    Parameters
    ----------
    ne : (nt, nx) array – electron density
    x  : (nx,)   array – spatial axis
    t  : (nt,)   array – time axis
    smooth_pts : int – boxcar smoothing width applied to each profile before
                 computing the gradient (set 1 to disable)
    threshold_frac : float – ignore gradients below this fraction of the
                     profile-maximum gradient (removes noise)
    max_jump : float or None – if set, constrain the front position to move
               at most this far between consecutive timesteps

    Returns
    -------
    dict  with keys "t", "x_front", "v_front"
    """
    ne, x, t = np.asarray(ne, float), np.asarray(x, float), np.asarray(t, float)
    nt, nx = ne.shape
    dx = np.median(np.diff(x))

    x_front = np.full(nt, np.nan)
    prev = None

    for it in range(nt):
        profile = ne[it]
        if smooth_pts > 1:
            profile = uniform_filter1d(profile, size=smooth_pts)

        grad = np.gradient(profile, dx)        # dne/dx
        grad_pos = np.where(grad > 0, grad, 0) # keep only rising edges

        gmax = grad_pos.max()
        if gmax <= 0:
            if prev is not None:
                x_front[it] = prev
            continue

        # zero out tiny gradients (noise)
        grad_pos[grad_pos < threshold_frac * gmax] = 0

        # candidate positions where gradient is large
        candidates = np.where(grad_pos > 0)[0]

        if prev is None:
            # first frame: pick leftmost strong gradient
            chosen = candidates[0]
        else:
            # pick candidate closest to previous front
            dists = np.abs(x[candidates] - prev)
            if max_jump is not None:
                ok = dists <= max_jump
                if ok.any():
                    candidates = candidates[ok]
                    dists = dists[ok]
            chosen = candidates[np.argmin(dists)]

        x_front[it] = x[chosen]
        prev = x_front[it]

    # velocity from finite differences
    v_front = np.full_like(x_front, np.nan)
    ok = np.isfinite(x_front)
    if ok.sum() >= 2:
        v_front[ok] = np.gradient(x_front[ok], t[ok])

    return {"t": t, "x_front": x_front, "v_front": v_front}


# ── Smilei helpers ──────────────────────────────────────────────────────────

def extract_ne(S, diag_number=3, y_halfwidth=5, units=("um", "fs", "cm^-3")):
    """Extract ne(t,x), x, t from a Smilei ParticleBinning diagnostic."""
    diag = S.ParticleBinning(
        diag_number, units=list(units),
        average={"z": "all",
                 "y": [int(S.namelist.Ly/2.) - y_halfwidth,
                       int(S.namelist.Ly/2.) + y_halfwidth]},
    )
    ne = np.array(diag.getData())
    t  = np.array(diag.getTimes())
    ax = np.array(diag.getAxis("x"))
    if ax.ndim == 2:
        ax = ax[:, 0]
    return ne, ax, t


# ── run multiple cases & plot ───────────────────────────────────────────────

def run_cases(S_list, a0_list, save_prefix=None, **tracker_kw):
    """
    Track density front for several Smilei runs and plot results.

    Extra keyword arguments are forwarded to track_front().
    """
    fig1, ax1 = plt.subplots(figsize=(8, 6))
    fig2, ax2 = plt.subplots(figsize=(8, 6))
    results = {}

    for S, a0 in zip(S_list, a0_list):
        ne, x, t = extract_ne(S)
        res = track_front(ne, x, t, **tracker_kw)
        results[a0] = res

        ax1.plot(res["t"], res["x_front"], "o-", ms=3, label=rf"$a_0={a0}$")
        ax2.plot(res["t"], res["v_front"], "o-", ms=3, label=rf"$a_0={a0}$")

    for ax, ylabel, title in [
        (ax1, "Front position (µm)", "Density-front position vs time"),
        (ax2, "Front velocity (µm/fs)", "Density-front velocity vs time"),
    ]:
        ax.set_xlabel("t (fs)")
        ax.set_ylabel(ylabel)
        ax.set_title(title)
        ax.legend()
        ax.grid(True, alpha=0.3)

    fig1.tight_layout()
    fig2.tight_layout()

    if save_prefix:
        fig1.savefig(f"{save_prefix}_front_position_vs_time.pdf",
                     dpi=300, bbox_inches="tight")
        fig2.savefig(f"{save_prefix}_front_velocity_vs_time.pdf",
                     dpi=300, bbox_inches="tight")

    return results
```

```python
results = run_cases(
    [S0, S1, S2, S3, S4, S5, S6],
    [300, 400, 500, 600, 700, 800, 900],
    save_prefix="ne_front",
    smooth_pts=9,          # adjust if profiles are noisy
    threshold_frac=0.1,    # raise if you still pick up noise
)
```

```python
# the above results are a bit messy for velocity after 100 fs. 
# this one is supposed to be better, with higher order of smoothing.

"""
Track the density front (steepest positive gradient dn/dx) instead of the
density peak.  More robust for strong-laser cases where multiple peaks form.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter1d


# ── core tracker ────────────────────────────────────────────────────────────

def track_front(ne, x, t, smooth_pts=5, threshold_frac=0.1, max_jump=None,
                v_smooth_pts=7):
    """
    At each timestep find the position of the largest positive dne/dx
    (i.e. the steepest rising edge = the shock / compression front).

    Parameters
    ----------
    ne : (nt, nx) array – electron density
    x  : (nx,)   array – spatial axis
    t  : (nt,)   array – time axis
    smooth_pts : int – boxcar smoothing width applied to each profile before
                 computing the gradient (set 1 to disable)
    threshold_frac : float – ignore gradients below this fraction of the
                     profile-maximum gradient (removes noise)
    max_jump : float or None – if set, constrain the front position to move
               at most this far between consecutive timesteps
    v_smooth_pts : int – boxcar smoothing width for the velocity curve (1=off)

    Returns
    -------
    dict  with keys "t", "x_front", "v_front"
    """
    ne, x, t = np.asarray(ne, float), np.asarray(x, float), np.asarray(t, float)
    nt, nx = ne.shape
    dx = np.median(np.diff(x))

    x_front = np.full(nt, np.nan)
    prev = None

    for it in range(nt):
        profile = ne[it]
        if smooth_pts > 1:
            profile = uniform_filter1d(profile, size=smooth_pts)

        grad = np.gradient(profile, dx)        # dne/dx
        grad_pos = np.where(grad > 0, grad, 0) # keep only rising edges

        gmax = grad_pos.max()
        if gmax <= 0:
            if prev is not None:
                x_front[it] = prev
            continue

        # zero out tiny gradients (noise)
        grad_pos[grad_pos < threshold_frac * gmax] = 0

        # candidate positions where gradient is large
        candidates = np.where(grad_pos > 0)[0]

        if prev is None:
            # first frame: pick leftmost strong gradient
            chosen = candidates[0]
        else:
            # pick candidate closest to previous front
            dists = np.abs(x[candidates] - prev)
            if max_jump is not None:
                ok = dists <= max_jump
                if ok.any():
                    candidates = candidates[ok]
                    dists = dists[ok]
            chosen = candidates[np.argmin(dists)]

        x_front[it] = x[chosen]
        prev = x_front[it]

    # velocity from finite differences, then smooth
    v_front = np.full_like(x_front, np.nan)
    ok = np.isfinite(x_front)
    if ok.sum() >= 2:
        v_raw = np.gradient(x_front[ok], t[ok])
        if v_smooth_pts > 1 and len(v_raw) >= v_smooth_pts:
            v_raw = uniform_filter1d(v_raw, size=v_smooth_pts)
        v_front[ok] = v_raw

    return {"t": t, "x_front": x_front, "v_front": v_front}


# ── Smilei helpers ──────────────────────────────────────────────────────────

def extract_ne(S, diag_number=3, y_halfwidth=5, units=("um", "fs", "cm^-3")):
    """Extract ne(t,x), x, t from a Smilei ParticleBinning diagnostic."""
    diag = S.ParticleBinning(
        diag_number, units=list(units),
        average={"z": "all",
                 "y": [int(S.namelist.Ly/2.) - y_halfwidth,
                       int(S.namelist.Ly/2.) + y_halfwidth]},
    )
    ne = np.array(diag.getData())
    t  = np.array(diag.getTimes())
    ax = np.array(diag.getAxis("x"))
    if ax.ndim == 2:
        ax = ax[:, 0]
    return ne, ax, t


# ── run multiple cases & plot ───────────────────────────────────────────────

def run_cases(S_list, a0_list, save_prefix=None, **tracker_kw):
    """
    Track density front for several Smilei runs and plot results.

    Extra keyword arguments are forwarded to track_front().
    """
    # well-separated colors for up to 7 curves
    COLORS = ["#d62728", "#e67e22", "#2ca02c", "#9467bd",
              "#1f77b4", "#17becf", "#333333"]

    fig1, ax1 = plt.subplots(figsize=(8, 6))
    fig2, ax2 = plt.subplots(figsize=(8, 6))
    results = {}

    for i, (S, a0) in enumerate(zip(S_list, a0_list)):
        ne, x, t = extract_ne(S)
        res = track_front(ne, x, t, **tracker_kw)
        results[a0] = res

        c = COLORS[i % len(COLORS)]
        ax1.plot(res["t"], res["x_front"], "o-", ms=3, color=c,
                 label=rf"$a_0={a0}$")
        # convert µm/fs → m/s  (1 µm/fs = 1e9 m/s)
        v_si = res["v_front"] * 1e9
        ax2.plot(res["t"], v_si, "o-", ms=3, color=c,
                 label=rf"$a_0={a0}$")

    for ax, ylabel, title in [
        (ax1, r"Front position ($\mu$m)", "Density-front position vs time"),
        (ax2, "Front velocity (m/s)", "Density-front velocity vs time"),
    ]:
        ax.set_xlabel("t (fs)")
        ax.set_ylabel(ylabel)
        ax.set_title(title)
        ax.legend()
        ax.grid(True, alpha=0.3)

    fig1.tight_layout()
    fig2.tight_layout()

    if save_prefix:
        fig1.savefig(f"{save_prefix}_front_position_vs_time.pdf",
                     dpi=300, bbox_inches="tight")
        fig2.savefig(f"{save_prefix}_front_velocity_vs_time.pdf",
                     dpi=300, bbox_inches="tight")

    return results
```
