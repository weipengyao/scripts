```python
## check the Ex field spatial distribution at the middle of the box

Ex = S2.Probe(0,'Ex',units=['V/m', 'um', 'fs'], vsym=True,
        #  subset={'axis2':int(S2.namelist.Ly/2.)},
         average={'axis2':[int(S2.namelist.Ly/2.)-5,int(S2.namelist.Ly/2.)+5]},
         ).slide()
```

```python
x = S2.Probe(0,'Ex',units=['um']).getAxis('axis1')[:,0]
idx_target = np.abs(x - 26).argmin()
print(x[idx_target])
print(idx_target) # this is the index of the target rear surface
idx_left_boundary = np.abs(x - 0.5).argmin()
print(x[idx_left_boundary])
print(idx_left_boundary) # this is the index of the left boundary
idx_target_front = np.abs(x - 6).argmin()
print(x[idx_target_front])
print(idx_target_front) # this is the index of the target front surface
```

```python
## Then, we need to find the position of the maximum Ex field, at every time step.
def find_max_Ex_position(S):
     time = S.Probe(0,'Ex',units=['fs']).getTimes()  # time array in fs
     x = S.Probe(0,'Ex',units=['um']).getAxis('axis1')[:,0] # x array in um

     x_pos = np.zeros(np.size(time)) # array to store the x position of the maximum Ex field at each time step

     for i in range(np.size(time)-1):
          i=i+1
          Ex = np.array(S.Probe(0,'Ex',units=['V/m'],
                                # subset={'axis2':int(S.namelist.Ly/2.)},
                                average={'axis2':[int(S2.namelist.Ly/2.)-10,int(S2.namelist.Ly/2.)+10]}, # do a bit average
                                ).getData()[i])
          max_pos = Ex[idx_left_boundary:idx_target].argmax()
          # max_pos = np.where(Ex == np.max(Ex))
          x_pos[i] = x[max_pos]
          print(x_pos[i])
          # if x_pos[i] > 25: 
          #      x_pos[i:] = np.nan # if the position is larger than 25 um, set it to nan (not a number) to ignore it in the plot
          #      break
     return time, x_pos
```

```python
time_a0300, x_pos_a0300 = find_max_Ex_position(S0)
time_a0400, x_pos_a0400 = find_max_Ex_position(S1)
time_a0500, x_pos_a0500 = find_max_Ex_position(S2)
time_a0600, x_pos_a0600 = find_max_Ex_position(S3)
time_a0700, x_pos_a0700 = find_max_Ex_position(S4)
time_a0800, x_pos_a0800 = find_max_Ex_position(S5)
time_a0900, x_pos_a0900 = find_max_Ex_position(S6)
```

```python
plt.figure(figsize=(10,8))
plt.plot(time_a0300, x_pos_a0300, label='a0=300', color='tab:blue')
plt.plot(time_a0400, x_pos_a0400, label='a0=400', color='tab:orange')
plt.plot(time_a0500, x_pos_a0500, label='a0=500', color='tab:green')
plt.plot(time_a0600, x_pos_a0600, label='a0=600', color='tab:red')
plt.plot(time_a0700, x_pos_a0700, label='a0=700', color='tab:purple')
plt.plot(time_a0800, x_pos_a0800, label='a0=800', color='tab:brown')
plt.plot(time_a0900, x_pos_a0900, label='a0=900', color='tab:pink')
plt.hlines(x[idx_target], time_a0400[0], time_a0400[-1], colors='k', linestyles='dashed', alpha=0.5, label='target rear surface')
plt.hlines(x[idx_target_front], time_a0400[0], time_a0400[-1], colors='k', linestyles='dashed', alpha=0.5, label='target front surface')
plt.xlim(0, 150)
plt.xlabel('Time (fs)')
plt.ylabel('Position of max Ex (um)')
plt.title('Position of maximum Ex field over time, with ne=120 nc')
plt.grid(alpha=0.3)
plt.legend(loc='best')
plt.savefig('/Users/yao/Desktop/max_Ex_position_ne120.png', dpi=300)
```
