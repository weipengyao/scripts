```python
def get_S_para_t(S, label=None):
    Ey = np.array(S.Probe(0,'Ey',units=["um", "fs"]).getData())
    S_para_t = np.zeros(Ey.shape[0])
    for i in range(Ey.shape[0]):
        j=i*5
        Ey = np.array(S.Probe(0,'Ey',units=["um", "fs"
                                            # , "V/m"
                                            ]).getData()[i])
        ne = np.array(S.ParticleBinning(3,units=["um", "fs"
                                                    # , 'C m^{-3}'
                                                    ]).getData()[j])
        # S_para_t[i] = np.sum(rho_e/nr/qe * (-1.0) * np.abs(Ey/Er)**2) / lmd
        S_para_t[i] = np.sum(ne[:,610,1] * np.abs(Ey[:,610])**2)

    time = np.array(S.Probe(0,'Ey',units=["um", "fs"]).getTimes())
    return time, S_para_t, label
```

```python
time0, S_para_t0, label0 = get_S_para_t(S0, 'a0=212')
time1, S_para_t1, label1 = get_S_para_t(S1, 'a0=282')
time2, S_para_t2, label2 = get_S_para_t(S2, 'a0=353')
time3, S_para_t3, label3 = get_S_para_t(S3, 'a0=424')
time4, S_para_t4, label4 = get_S_para_t(S4, 'a0=494')
time5, S_para_t5, label5 = get_S_para_t(S5, 'a0=565')
time6, S_para_t6, label6 = get_S_para_t(S6, 'a0=636')
```
