```python
def energy_budget(S):
    utot = S.Scalar("Utot", units=["J", 'fs'], label="Total energy")
    ukin = S.Scalar("Ukin", units=["J", 'fs'], label="Kinetic energy")
    uelm = S.Scalar("Uelm", units=["J", 'fs'], label="Electromagnetic energy")
    urad = S.Scalar("Urad", units=["J", 'fs'], label="Radiation energy")
    happi.multiPlot(utot, ukin, uelm, urad)
    return 0

def radiation(S, a0):
    urad = S.Scalar("Urad", units=["J", 'fs'], label=a0, 
                    # data_log=True
                    )
    return urad

def energy_quantify(S):
    utot_max = np.max(S.Scalar("Utot", units=["J", 'fs']).getData())
    urad_max = np.max(S.Scalar("Urad", units=["J", 'fs']).getData())
    ueon_max = np.max(S.Scalar("Ukin_eon", units=["J", 'fs']).getData())
    rad_eff = urad_max / utot_max
    eta_rad = ueon_max / utot_max
    print(f"Utot_max: {utot_max:.2e} J")
    print(f"Urad_max: {urad_max:.2e} J")
    print(f"Ueon_max: {ueon_max:.2e} J")
    print(f"Radiation efficiency (from Urad): {rad_eff:.2%}")
    print(f"Radiation efficiency (from Ukin_eon): {eta_rad:.2%}")

    
# utot = S0.Scalar("Utot", units=["J", 'fs'], label="Total energy")
# ukin = S0.Scalar("Ukin", units=["J", 'fs'], label="Kinetic energy")
# uelm = S0.Scalar("Uelm", units=["J", 'fs'], label="Electromagnetic energy")
# urad = S0.Scalar("Urad", units=["J", 'fs'], label="Radiation energy")

# happi.multiPlot(utot, ukin, uelm, urad)
```

```python
a0 = np.array([300, 400, 500, 600, 700], dtype=float)

# original definition of radiation efficiency, using radiation energy, which is directly available in the simulation outputs. The original values of radiation efficiency are calculated as the ratio of the maximum radiation energy to the maximum total energy, and are expressed as percentages for better readability.
rad_eff = np.array([0.18, 0.91, 3.23, 7.09, 10.87], dtype=float) 

# updated definition of radiation efficiency, using kinetic energy of electrons instead of radiation energy, since the latter is not directly available in the simulation outputs. The new definition is based on the assumption that the kinetic energy of electrons can be a proxy for the energy lost to radiation, which is reasonable in the ultra-relativistic regime where radiation reaction is significant. The new values of radiation efficiency are calculated as the ratio of the maximum kinetic energy of electrons to the maximum total energy, and are expressed as percentages for better readability.
# rad_eff = np.array([3.45, 6.32, 8.68, 10.17, 10.58], dtype=float) 

plt.figure(figsize=(6, 5))
plt.plot(a0, rad_eff, marker="o", markersize=8,linestyle="none", color="r")
plt.xlabel("a0")
plt.ylabel(r"$\eta_{\rm rad}$ (%)")
# plt.title("Radiation efficiency vs a0")
plt.grid(which="both", linestyle="--", linewidth=0.5)
# fitting the data with a power law
def power_law(x, a, b):
    return a * x**b 
from scipy.optimize import curve_fit
popt, pcov = curve_fit(power_law, a0, rad_eff)
a_fit, b_fit = popt
print(f"Fitted power law: efficiency = {a_fit:.2e} * a0^{b_fit:.2f}")
x_fit = np.linspace(300, 700, 100)
y_fit = power_law(x_fit, a_fit, b_fit)
plt.plot(x_fit, y_fit, label=f"Fit: {a_fit:.2e} * a0^{b_fit:.2f}")
plt.legend()
plt.show()
plt.savefig("/Users/yao/Desktop/ife_radiation_efficiency2.pdf", dpi=300)
```

```python
U_SI = 1.

def absorption(S):
    Pin = np.array(S.Scalar('PoyXmin',units=["J",'fs']).getData())

    Pout = np.abs(np.array(S.Scalar('PoyXmax',units=["J",'fs']).getData())) + np.abs(np.array(S.Scalar('PoyYmin',units=["J",'fs']).getData())) + np.abs(np.array(S.Scalar('PoyYmax',units=["J",'fs']).getData())) +  np.abs(np.array(S.Scalar('PoyZmin',units=["J",'fs']).getData())) + np.abs(np.array(S.Scalar('PoyZmax',units=["J",'fs']).getData()))

    Uelm_Bx = np.array(S.Scalar('Uelm_Bx_m', units=["J"]).getData())
    Uelm_By = np.array(S.Scalar('Uelm_By_m', units=["J"]).getData())
    Uelm_Bz = np.array(S.Scalar('Uelm_Bz_m', units=["J"]).getData())

    Uelm_Bx_J = Uelm_Bx[-1]
    Uelm_By_J = Uelm_By[-1]
    Uelm_Bz_J = Uelm_Bz[-1]

    LASER_ENERGY_UTOT_J = np.max(S.Scalar('Utot',units=["J"]).getData())
    LASER_ENERGY_POY_J = Pin[-1]

    PLASMA_ENERGY_POY_J = (Pin[-1]-Pout[-1])
    PLASMA_ENERGY_UKIN_J = np.max(S.Scalar('Ukin',units=["J"]).getData())




    B_ENERGY = Uelm_Bx_J+Uelm_By_J+Uelm_Bz_J

    Bx_ENERGY = Uelm_Bx_J

    print("Laser energy (Utot and Poy):",round(LASER_ENERGY_UTOT_J,3),"J and",round(LASER_ENERGY_POY_J,3),"J")
    print("Plasma energy (Ukin and Poy) :",round(PLASMA_ENERGY_UKIN_J,3),"J and",round(PLASMA_ENERGY_POY_J,3),"J")
    print("Plasma ABS: (U and Poy)", round(PLASMA_ENERGY_UKIN_J/LASER_ENERGY_UTOT_J*100,3),"% and", round(PLASMA_ENERGY_POY_J/LASER_ENERGY_POY_J*100,3),"%")
    print("=====================")

    print("Magnetic field energy:",round(B_ENERGY,5),"J")
    print("Magnetic field ABS:",round(B_ENERGY/PLASMA_ENERGY_POY_J*100,3),"%")

    print("Bx energy:",round(Bx_ENERGY,5),"J")
    print("Bx ABS: (U and Poy)",round(Bx_ENERGY/PLASMA_ENERGY_UKIN_J*100,3),"% and", round(Bx_ENERGY/PLASMA_ENERGY_POY_J*100,3),"%")


    print("=====================")
    print("Total ABS: (U and Poy)", round(Bx_ENERGY/LASER_ENERGY_UTOT_J*100,4),"% and", round(Bx_ENERGY/LASER_ENERGY_POY_J*100,4),"%")
```
