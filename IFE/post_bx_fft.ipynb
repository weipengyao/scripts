```python
data = np.array(Bx.getData()[-1])  # get the field data at timestep 1456
```

```python
field_x = np.array(Bx.getAxis("axis1"))
field_y = np.array(Bx.getAxis("axis2"))
```

```python
Nx = data.shape[0]
Ny = data.shape[1]
print(Nx, Ny)
```

```python
dx = (field_x[1] - field_x[0])[0]
dy = (field_y[1] - field_y[0])[1]
print(dx, dy)
```

```python
kx = np.fft.fftfreq(Nx, d=dx) #/ (2 * np.pi)  # spatial freq extents
ky = np.fft.fftfreq(Ny, d=dy) #/ (2 * np.pi)
KX, KY = np.meshgrid(kx, ky, indexing="ij")
```

```python
data_fft = np.fft.fft2(data, axes=(0, 1))  # shape (Nx, Ny)

plt.imshow(np.log10(np.abs(data_fft).T), extent=(kx.min(), kx.max(), ky.min(), ky.max()), origin="lower")
plt.colorbar()
plt.xlabel("kx (1/um)")
plt.ylabel("ky (1/um)")
plt.title("Fourier Transform of By in log10 scale")
```

```python
data_fft_shift = np.fft.fftshift(data_fft)

plt.imshow(np.log10(np.abs(data_fft_shift).T), extent=(kx.min(), kx.max(), ky.min(), ky.max()), origin="lower")
plt.colorbar()
plt.xlabel("kx (1/um)")
plt.ylabel("ky (1/um)")
plt.title("Shifted Fourier Transform of By in log10 scale")
plt.xlim(-2,2)
plt.ylim(-2,2)
```

```python
pwr = 12  # super gaussian power

# adjust the mask parameters here according to the FFT plot
k0x = 0.4
k0y = 0.4
wx = 0.3
wy = 0.4

mask = 1-np.exp(-(np.sqrt(((np.abs(KX) - k0x) / wx) ** 2 + ((np.abs(KY) - k0y) / wy) ** 2) ** pwr))
# a quick check of the mask
plt.imshow(mask.T, extent=(kx.min(), kx.max(), ky.min(), ky.max()), origin="lower")
plt.colorbar()
plt.xlabel("kx (1/um)")
plt.ylabel("ky (1/um)")
plt.title("Fourier Mask")
```

```python
mask_shift = np.fft.fftshift(mask)
plt.imshow(mask_shift.T, extent=(kx.min(), kx.max(), ky.min(), ky.max()), origin="lower")
plt.colorbar()
plt.xlabel("kx (1/um)")
plt.ylabel("ky (1/um)")
plt.title("Shifted Fourier Mask")
plt.xlim(-2,2)
plt.ylim(-2,2)
```

```python
data_fft_masked = data_fft * mask
data_fft_masked_shift = np.fft.fftshift(data_fft_masked)
plt.imshow(np.log10(np.abs(data_fft_masked_shift).T), extent=(kx.min(), kx.max(), ky.min(), ky.max()), origin="lower")
plt.colorbar()
plt.xlabel("kx (1/um)")
plt.ylabel("ky (1/um)")
plt.title("Masked & Shifted Fourier Transform of By in log10 scale")
plt.xlim(-2,2)
plt.ylim(-2,2)
```

```python
Diag_data = np.real(np.fft.ifft2(data_fft_masked))

plt.imshow(Diag_data.T, 
           extent=(field_x.min(), field_x.max(), field_y.min(), field_y.max()), 
           origin="lower",
           cmap='bwr',
           aspect='auto',
           vmin=-2,
           vmax=2,
           )
plt.colorbar()
plt.xlabel("x (um)")
plt.ylabel("y (um)")
plt.title("Filtered Bx field in real space")
```

```python
xx = np.linspace(field_x.min(), field_x.max(), Nx)
plt.plot(xx, data[:, Ny//2], label='Original Bx')
plt.plot(xx, Diag_data[:, Ny//2], label='Filtered Bx')
plt.xlabel("x (um)")
plt.ylabel("Bx (1e5 T)")
plt.title("Lineout of Bx at y=0")
plt.legend()
```
