```python
Lx = 60

x = np.linspace(0, Lx, 1000)

plasma_ramp = 1.
ne = 1.0e22

def fne_plasma(x):
    # return ne*np.exp( -(x-Lx/2)**12/(Lx/2-plasma_ramp)**12 )
    return ne*np.exp( -(x-Lx/2)**12/(1.5-plasma_ramp)**12 )

plt.plot(x, fne_plasma(x))
```
