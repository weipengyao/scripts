```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Nov 24 12:00:51 2022

@author: yao
"""

import matplotlib.pyplot as plt

# Publication Quality Figure Parameters
plt.rc('font', family='sans-serif', serif='Arial')   # Spend a long time dealing with this. 
# plt.rc('font', family='Liberation Mono')
plt.rc('text', usetex=False)
plt.rc('xtick', labelsize=16)
plt.rc('ytick', labelsize=16)
plt.rc('axes', labelsize=16)
plt.rc('savefig', dpi=600)

width = 3.487 * 2.0# width of an one-column journal, for yz-plane
height = width * 1.618 / 1.0# golden ratio

# fig. 1b for cnrs concours 2023

years_tnsa = [2000,  2016]
pmax_tnsa  = [58.0,  85.0]

years_rpa  = [2016, 2018, 2022]
pmax_rpa   = [93.0, 94.0, 150.]

years_goal = [2026, 2028]
pmax_goal  = [500,  1000]

years = [2000, 2011, 2016, 2018, 2022, 2025]
pmaxs = [58.0, 67.5, 93.0, 94.0, 150., 700]

# fig, ax = plt.figure(figsize=(width, height))
plt.plot(years_tnsa, pmax_tnsa, 'ok',markersize=10, fillstyle='full', label='TNSA')
plt.plot(years_rpa,  pmax_rpa,  '^b',markersize=12, fillstyle='full', label='hybrid TNSA/RPA')
plt.plot(years_goal, pmax_goal, 'D-r',markersize=10, lw=3.0, label='RPA at Apollon')
# plt.vlines(years_goal, ymin=500, ymax=1000, label='RPA at Apollon', linewidth=3.0, color='red')
plt.xlabel('Years',fontsize=20)
plt.ylabel('Max. proton energy (MeV)',fontsize=20)
plt.yscale('log')
plt.legend(frameon=False,fancybox=False,fontsize=16, loc='upper left')
plt.grid(which='both',linestyle='--',alpha=0.5)

plt.xlim([1999, 2030])

plt.text(1999, 65, '[SNA]',fontsize=16)
# plt.text(2009, 75, '[GAI]',fontsize=16)
plt.text(2015, 65, '[WAG]',fontsize=16)
plt.text(2014, 100, '[KIM]',color='blue', fontsize=16)
plt.text(2018, 102, '[HIGa]',color='blue', fontsize=16)
plt.text(2022, 165, '[ZIE]',color='blue', fontsize=16)
plt.text(2021.5, 460, '2 PW',color='red', fontsize=16)
plt.text(2022.5, 930, '10 PW',color='red', fontsize=16)
plt.tight_layout()
# plt.show()

plt.savefig('/Users/yao/Desktop/fig1b.pdf', format='pdf')
```
