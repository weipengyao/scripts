```python
Urad = np.array(S1.Scalar("Urad",
                        #   units=['J/um']
                          ).getData())
```

```python
Urad_time = np.array(S1.Scalar("Urad",
                            #    units=['J/um','s']
                               ).getTimes())
dt = Urad_time[2] - Urad_time[1]  # in s
```

```python
P_Scalar = (Urad[43] - Urad[42])/dt * 1e-4 # J/s/um in 1 um
print('Power from Scalar is {:.2e} W'.format(P_Scalar))
```

```python
N = Urad.size
P_Scalar = np.zeros(N)
P_Bin    = np.zeros(N)

for i in range(N-1):
    P_Scalar[i] = (Urad[i+1] - Urad[i]) / dt
    P_Bin[i] = (dgamma[2] - dgamma[1]) * np.sum(P_rad[i,:]) 

```

```python
plt.plot(Urad_time, P_Scalar,'-r',label='Scalar')
plt.plot(Urad_time, P_Bin*7e4,'--b',label='Binning * 7e4')
plt.legend()
```
